```python
#pattern1
for i in range(5):
    for j in range(5):
        print("*",end=" ")
    print()
```

```python
#pattern2
for i in range(6):
    print("*"*i,end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(i+1):
        print("*",end=" ")
    print()
```

```python
#pattern3
n=5
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
#pattern4
n=5
for i in range(1,n+1):
    for j in range(1,i+1):
        print(i,end=" ")
    print()
```

```python
#pattern5
for i in range(5):
    for j in range(5,i,-1):
        print(j,end=" ")
    print()
```

```python
#pattern6
for i in range(5):
    for j in range(5,i,-1):
        print(5-j+1,end=" ")
    print()
```

```python
#pattern7
for i in range(5):
    for j in range(5,i,-1):
        print("*",end=" ")
    print()
```

```python
#pattern8
n=5
for i in range(n):
    for j in range(n-i-1):
        print(" ",end=" ")
    for j in range(2*i+1):
        print("*",end=" ")
    for j in range(n-i-1):
        print(" ",end=" ")
    print()
```
